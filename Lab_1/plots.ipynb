```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
model_5 = pd.read_csv('5_model/model_5.csv')
model_20 = pd.read_csv('20_model/model_20.csv')
model_50 = pd.read_csv('50_model/model_50.csv')
model_100 = pd.read_csv('100_model/model_100.csv')
model_300 = pd.read_csv('300_model/model_300.csv')
```

```python
model_5.head()
```

```python
#Add a title to the main figure


plt.figure(figsize=(10, 5))
                          

plt.subplot(1, 2, 1)
plt.plot(model_5['accuracy'], label='5 Neurons')
plt.plot(model_20['accuracy'], label='20 Neurons')
plt.plot(model_50['accuracy'], label='50 Neurons')
plt.plot(model_100['accuracy'], label='100 Neurons')
plt.plot(model_300['accuracy'], label='300 Neurons')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plotting loss
plt.subplot(1, 2, 2)
plt.plot(model_5['loss'][1:], label='5 Neurons')
plt.plot(model_20['loss'][1:], label='20 Neurons')
plt.plot(model_50['loss'][1:], label='50 Neurons')
plt.plot(model_100['loss'][1:], label='100 Neurons')
plt.plot(model_300['loss'][1:], label='300 Neurons')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('plots.png', dpi=300)
```
